```python
import pandas as pd
import numpy as np
import strsimpy
print('setup done!')
```

```python
df = pd.read_json('./data.json')
arr = df.values.tolist()

levenshtein  = strsimpy.NormalizedLevenshtein()

def levenshtein_score(x,y):
    if (x=='unknown' or y =='unknown'):
        return 1
    return 1 - levenshtein.distance(x,y)

def difference_score(x,y):
    if (x=='unknown' or y =='unknown'):
        return 1
    return 1 if x==y else 0
```

```python
def clust_filter(arr):
    flag = [False]*len(arr)
    cluster  = []
    for i in range(0,len(arr)-1):
        if (flag[i]==False):
            list_id = []
            for j in range(i + 1, len(arr)):
                if (flag[j]==False):
                    name = levenshtein_score(arr[i][0],arr[j][0])
                    cpu = difference_score(arr[i][2],arr[j][2])
                    ram = difference_score(arr[i][5],arr[j][5])
                    rom = difference_score(arr[i][6],arr[j][6])
                    color = difference_score(arr[i][10],arr[j][10])
                    score = 0.5 * name + 0.05 * cpu + 0.2 * ram + 0.2 * rom + 0.05 * color
                    if (score > 0.95):
                        flag[j] = True
                        list_id.append(j)
            arr[i].append(list_id)
            cluster.append(arr[i])
    return cluster
```

```python
cluster = clust_filter(arr)
```

```python
new_df = pd.DataFrame(cluster, columns=['name','url', 'cpu', 'os', 'gpu', 'ram', 'rom', 'price', 'connect', 'screentype',
 'color', 'front_camera', 'rear_camera', 'battery_capacity', 'bluetooth',
 'wifi', 'size', 'resolution', 'weight', 'jack_audio', 'mobile_network', 'sim',
 'security', 'material','list_id'])
# print(len(new_df))
print(new_df)
```
