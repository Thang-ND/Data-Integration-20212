```python
from SchemaMatchingSystem import SchemaMatchingSystem
import pandas as pd
```

```python
df = pd.read_json("../PreprocessingData/Dataprocess/iphone/thegioididong.json").rename(columns={"Màn hình": "size"})
df.head(5)
```

```python
matchingSystem = SchemaMatchingSystem()
columns = list(df.columns)
```

```python
res = matchingSystem.schema_matching(cols=columns)
res
```

```python
target_schema = matchingSystem.getTargetSchema()
data = pd.DataFrame(columns=target_schema)
```

```python
def matchedCol(result): 
    matched_col = []
    source_col = []
    for key in result:
        matched_col.append(result[key])
        source_col.append(key)
    
    return matched_col, source_col

matched_col, source_col = matchedCol(res)
```

```python
df_tmp = df[source_col].rename(columns=res)
data[matched_col] = df_tmp[matched_col]
data
```

```python
data.to_json("./IntegratedData/data4.json")
```
