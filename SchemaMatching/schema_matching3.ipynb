```python
from SchemaMatchingSystem import SchemaMatchingSystem
import pandas as pd
```

```python
df = pd.read_json("../PreprocessingData/Dataprocess/iphone/applescraper-applewatch.json")
df.head(5)
```

```python
df.info()
```

```python
df2 = df.drop(["Name","Price", "ROM_P", "Color", "Price_color"], axis=1)
df2 = df2.rename(columns={"Model-split":"name", "Color-split":"color", "Price-split":"price"})
```

```python
df2.info()
```

```python
columns = list(df2.columns[0:17])
df2 = df2[columns]
```

```python
matchingSystem = SchemaMatchingSystem()
res = matchingSystem.schema_matching(cols=columns)
```

```python
target_shema = matchingSystem.getTargetSchema()
data = pd.DataFrame(columns=target_shema)
data
```

```python
def matchedCol(result): 
    matched_col = []
    source_col = []
    for key in result:
        matched_col.append(result[key])
        source_col.append(key)
    return matched_col, source_col
```

```python
matched_col, source_col = matchedCol(res)
```

```python
df2 = df2[source_col].rename(columns=res)
data[matched_col] = df2[matched_col]
data
```

```python
data.to_json("./IntegratedData/data7.json")
```
